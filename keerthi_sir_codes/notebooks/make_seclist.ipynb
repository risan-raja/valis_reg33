```python
import requests
```

```python
bsids = ['141','222','142','213','244']
```

```python
dataarg = 3
```

```python

endpoint='http://ap7.humanbrain.in:8000'
authtuple=('admin','admin')

sslist = requests.get(endpoint+'/qc/SeriesSet/',auth=authtuple).json()
ssdict = {elt['biosample']:(elt['id'],elt['name']) for elt in sslist}
# self.ssid,self.bsname = ssdict[int(bsid)]

```

```python
bsid = bsids[dataarg]
```

```python
ssid = ssdict[int(bsid)]
```

```python
ssid
```

```python
tnail_data=requests.get(endpoint+'/GW/getBrainViewerDetails/IIT/V1/SS-%d?public=0' % ssid[0]).json()
nisl_data = tnail_data['thumbNail']['NISSL']
nisl_dict={elt['position_index']:elt for elt in nisl_data}
```

```python
nisl_dict.keys()
```

```python
annot_data=requests.get(endpoint+'/GW/getBrainThumbNailDetails/IIT/V1/SS-%d?public=1' % ssid[0]).json()
```

```python
annot_list = [elt['sectionNo'] for elt in annot_data]
```

```python
lookup = {sec: sec in annot_list for sec in nisl_dict}
```

```python
import json
json.dump(lookup, open('/data/keerthi/brainpubdata/seclist_%d.json' % dataarg,'wt'))
```
